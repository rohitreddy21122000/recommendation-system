```python
# Implementation of Movie Recommender System


# import pandas library
import pandas as pd

# Get the data of rating( read the data from csv file or excel)

ratings = pd.read_csv('ratings.tsv',sep='\t',names=['user_id', 'item_id', 'rating', 'timestamp'])

# get data of movie titles
titles = pd.read_csv('Movie_Id_Titles.csv')
```

```python
ratings.head()
```

```python
titles.head()
```

```python
data = pd.merge(ratings,titles,on='item_id')
data.head()
```

```python
# Calculate mean rating of all movies
x=data.groupby('title')['rating'].mean()        
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(20)
```

```python
#Calculate count rating of all movies
y=data.groupby('title')['rating'].count()
data.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
```

```python
# creating dataframe with 'rating' count values
data_frame1 = pd.DataFrame(x)
data_frame1['no_of_votes'] = y
data_frame1.sort_values(by = ['no_of_votes'],ascending=False).head(15)

```

```python
data_frame2 = data.pivot_table(index ='user_id',columns ='title',values ='rating')
data_frame2.head()
```

```python
# analysing correlation with similar movies
starwars_ratings = data_frame2['Star Wars (1977)']
liarliar_ratings = data_frame2['Liar Liar (1997)']
```

```python
#find out movies similar to starwars, liar liar 
similar_to_starwars = data_frame2.corrwith(starwars_ratings)
similar_to_liarliar = data_frame2.corrwith(liarliar_ratings)
#here similarity of movies is given by similarity_factor normalised bewteen 0 to 1
#movie with 1 as similarity_factor is very much similar to the starwars or liar liar
```

```python
#top five movies that are very similar to starwars
similar_to_starwars[similar_to_starwars>0.8].sort_values(ascending=True).head(5)
```

```python
#top five movies that are very similar to liarliar
similar_to_liarliar[similar_to_liarliar>0.8].head(5)
```
